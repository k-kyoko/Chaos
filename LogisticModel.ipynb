```julia
# purpose
# 指数的成長モデルとロジスティックモデル
# の写像の収束を比較するためにプロットする。
# f(x) = 2x
# g(x) = 2x(1-x)
```

```julia
# module import / Basic Strings
using Plots     # import namedefineは＠
gr()            # プロットの流派　一番軽いやつを選んでやってくれる　matplotlibに近い　普通　plotlyは対応してないけど重いけど便利　

using TypedTables
```

```julia
# 関数の宣言
# 実行をまとめたやつ　１行で書きたくないやつ　function asdf(x, y)
# かんたんなやつ function
f(x) = 2x 　　　# ２つ以上の変数をかけるのはかける必要
g(x) = 2x * (1-x)
```

```julia
# method
# 定数、条件設定
N = 12                     # 試行回数　初期値不考慮
x_0 = 0.01                 #初期値 LAtEXの場合下付き文字で出てくるから atomでもそう見える

# 軌道、x, f(x), f^k(x)を得る　
x = Vector{Float64}(N+1)   # Float64で小数の縛りを入れる
x[1] = x_0  #初期値は１から
# これで軌道を入れる器ができた

function trial(func, x, n)
    d = copy(x)
    for i = 1:n
        d[i+1] = func(d[i])
    end
    return d                # 省略可能
end
```

```julia
# result
f_a = trial(f, x, N)
g_a = trial(g, x, N)
x = [i for i=0:12] # x軸のやつ
; #消える
```

```julia
# plot
# 横軸は配列の数　
plot(x, [f_a, g_a],label=["f(x) = 2x", "g(x) = 2x(1-x)"], lw=3)   #　ラベルの命名、線の太さ
```
